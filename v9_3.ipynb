## Data Loading

```python
import pandas as pd

# Load the dataset
file_path = 'Input_MakeMyTrip.csv'
df = pd.read_csv(file_path, error_bad_lines=False)  # Loading the dataset, ignoring bad lines
```

## Basic Data Exploration

```python
# Displaying the first few rows of the dataset
df.head()  # View the first few rows for an initial assessment of data
```

```python
df.shape  # Check the total number of rows and columns
```

```python
df.info()  # Get detailed information on column data types and non-null values
```

```python
# Define columns to be dropped - these are not necessary for analysis and prediction as they are irrelevant or are empty
columns_to_drop_extended = ['Uniq Id', 'Crawl Timestamp', 'Page Url', 
                            'Flight Stops', 'Meals', 'Initial Payment For Booking', 'Date Change Rules', 'Cancellation Rules','Onwards Return Flight Time', 'Price Per Two Persons']


df_reduced = df.drop(columns_to_drop_extended, axis=1)
df_reduced.head()
```

```python
df_reduced.shape
```

```python
#find the irregularities in the data
df_reduced['Places Covered'].value_counts()
```

```python
#Drop the rows which start with a number
df_reduced = df_reduced[~df_reduced['Places Covered'].str.contains(r'[0-9]')]
df_reduced['Places Covered'].value_counts()
```

```python
#Check if values in places covered and destination are same
df_reduced['Places Covered'].equals(df_reduced['Destination'])
```

```python
#Drop the destination column as it is same as places covered
df_reduced = df_reduced.drop(['Destination'], axis=1)
```

```python
df_reduced.shape
```

```python
#Find missing values
df_reduced.isnull().sum()
```

```python
#get df of rows with missing values
df_missing = df_reduced[df_reduced.isnull().any(axis=1)]
df_missing.shape
```

```python
#drop rows with missing values
df_reduced = df_reduced.dropna()
```

```python
# Iterate through the df_missing DataFrame
for index, row in df_missing.iterrows():
    # Extract the package name
    package_name = row['Package Name']
    
    # Find the corresponding row in df_reduced
    matching_row = df_reduced[df_reduced['Package Name'] == package_name]
    
    # If a matching row is found
    if not matching_row.empty:
        # Iterate through columns to fill missing values
        for col in df_missing.columns:
            # Check if the value is missing in df_missing
            if pd.isna(row[col]):
                # Fill the missing value with the value from the matching row in df_reduced
                df_missing.at[index, col] = matching_row.iloc[0][col]
```

```python
df_missing.isnull().sum()
```

```python
# Concatenate df_reduced and df_missing
df_reduced = pd.concat([df_reduced, df_missing], ignore_index=True)
```

```python
#Fill the missing values with the mode
df_reduced['Airline'].fillna(df_reduced['Airline'].mode()[0], inplace=True)
df_reduced['Airline'].isnull().sum()
```

```python
df_reduced.isnull().sum()
```

```python
#drop rows where sightseeing is empty
df_reduced = df_reduced.dropna(subset=['Sightseeing Places Covered'])
```

```python
#fill mode for the missing values of Hotel Details
df_reduced['Hotel Details'].fillna(df_reduced['Hotel Details'].mode()[0], inplace=True)
```

```python
#copy the dataframe
df_reduced_cleaned = df_reduced.copy()
```

```python
# Verify that there are no more missing values
df_reduced_cleaned.isnull().sum()
```

## Feature engineeing


```python
# Analyzing the dataset for feature engineering and encoding categorical variables
df_reduced_cleaned.head()
```

Package Name

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# Visualize data to gain insights, such as using a word cloud for 'Package Name'

# Generating a word cloud for the 'Package Name' column
text = ' '.join(df_reduced_cleaned['Package Name'])

# Creating a WordCloud object
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(text)

# Plotting the WordCloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Extracting TF-IDF features from 'Package Name'
tfidf = TfidfVectorizer(max_features=50)  # Limiting to top 50 features for simplicity
tfidf_matrix = tfidf.fit_transform(df_reduced_cleaned['Package Name'])

# Creating a DataFrame from the TF-IDF features
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out(), index=df_reduced_cleaned.index)

# Displaying the shape of the TF-IDF DataFrame and a few rows to understand its structure
tfidf_df_shape = tfidf_df.shape
tfidf_df_head = tfidf_df.head()

tfidf_df_shape, tfidf_df_head
```

```python

df_tfidf = df_reduced_cleaned.join(tfidf_df)
# Dropping the original 'Package Name' column
df_tfidf = df_tfidf.drop('Package Name', axis=1)

df_tfidf.head()
```

Package type

```python
# Displaying the unique values in the 'Package Type' column
unique_package_types = df_tfidf['Package Type'].unique()
unique_package_types
```

```python
import seaborn as sns

# Plotting the distribution of package types
plt.figure(figsize=(10, 6))
sns.countplot(data=df_tfidf, x='Package Type', order = df_tfidf['Package Type'].value_counts().index)
plt.title('Distribution of Package Types')
plt.xlabel('Package Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Encoding the 'Package Type' column using one-hot encoding
# This is done to convert the categorical values into numerical values

# Creating an instance of the OneHotEncoder class

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore')

# Encoding the 'Package Type' column
encoded_package_type = encoder.fit_transform(df_tfidf[['Package Type']])
encoded_package_type
```

```python

# Creating a DataFrame from the encoded values
encoded_package_type_df = pd.DataFrame(encoded_package_type.toarray(), columns=encoder.get_feature_names_out(), index=df_tfidf.index)
encoded_package_type_df
```

```python
# Joining the encoded DataFrame with the original DataFrame
df_encoded = df_tfidf.join(encoded_package_type_df)
df_encoded.head()

# Dropping the original 'Package Type' column
df_encoded = df_encoded.drop('Package Type', axis=1)
```

Company

```python
# Displaying the unique values in the 'Company' column
unique_companies = df_encoded['Company'].unique()
unique_companies
```

```python
# Dropping the 'Company' column as it contains only one unique value
df_reduced_no_company = df_encoded.drop('Company', axis=1)

# Displaying the shape of the dataset after dropping the 'Company' column
df_reduced_no_company.shape
```

```python
df_reduced_no_company.head()
```

Itenary

```python
from sklearn.preprocessing import OneHotEncoder

 # Example: Extract 'Total Nights' from 'Itinerary' column and add it as a new column to the dataset by summing up the number of nights for each row
df_reduced_no_company['Total Nights'] = df_reduced_no_company['Itinerary'].str.extractall(r'(\d+)N').astype(int).groupby(level=0).sum()

#display the changes
df_reduced_no_company[['Itinerary', 'Total Nights']].head()

#dropping the 'Itinerary' column
df_reduced_no_company = df_reduced_no_company.drop('Itinerary', axis=1)
```

Places covered

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Preparing the data for multi-label binarization for 'Places Covered'
# Splitting the places into a list of places per row
places_covered_lists = df_reduced_no_company['Places Covered'].str.split('|')

# Initializing the MultiLabelBinarizer
mlb_places = MultiLabelBinarizer()

# Applying MultiLabelBinarizer to the places covered lists
places_covered_encoded = mlb_places.fit_transform(places_covered_lists)

# Creating a DataFrame from the encoded data
places_covered_encoded_df = pd.DataFrame(places_covered_encoded, columns=mlb_places.classes_, index=df_reduced_no_company.index)

# Prefixing the column names with "Covered_" to distinguish from destinations
places_covered_encoded_df.columns = ['Covered_' + col for col in places_covered_encoded_df.columns]

# Displaying the shape of the new DataFrame and a few rows to understand its structure
places_covered_encoded_df_shape = places_covered_encoded_df.shape
places_covered_encoded_df_head = places_covered_encoded_df.head()

places_covered_encoded_df_shape, places_covered_encoded_df_head
```

```python
# Merging the encoded places covered data back into the main dataset
df_final_with_places_covered = df_reduced_no_company.join(places_covered_encoded_df)

# Displaying the shape of the final dataset to understand its structure after merging
df_final_with_places_covered.shape
```

```python
# Dropping the original 'Places Covered' column
df_final_with_places_covered.drop('Places Covered', axis=1, inplace=True)
```

Travel date

```python
import datetime

# Converting 'Travel Date' to datetime
df_final_with_places_covered['Travel Date'] = pd.to_datetime(df_final_with_places_covered['Travel Date'], errors='coerce')

# Extracting year, month, and day
df_final_with_places_covered['Year'] = df_final_with_places_covered['Travel Date'].dt.year
df_final_with_places_covered['Month'] = df_final_with_places_covered['Travel Date'].dt.month
df_final_with_places_covered['Day'] = df_final_with_places_covered['Travel Date'].dt.day

# Extracting day of the week
df_final_with_places_covered['Day of Week'] = df_final_with_places_covered['Travel Date'].dt.dayofweek

# Displaying the first few rows of the dataset with the new date features
df_final_with_places_covered[['Travel Date', 'Year', 'Month', 'Day', 'Day of Week']].head()

# Dropping the original 'Travel Date' column
df_final_with_places_covered.drop('Travel Date', axis=1, inplace=True)
```

Hotel details

```python
import re

# Feature 1: Number of Hotels
# Counting the number of different hotels mentioned for each package
# Assuming each hotel is separated by a '|'
df_final_with_places_covered['Number of Hotels'] = df_final_with_places_covered['Hotel Details'].str.count('\|') + 1

# Feature 2: Average Hotel Rating
# Extracting hotel ratings and calculating the average rating
def extract_average_rating(hotel_details):
    # Finding all occurrences of hotel ratings (assumed to be in the format 'Hotel Name:Rating')
    ratings = re.findall(r':(\d\.\d)', hotel_details)
    ratings = [float(rating) for rating in ratings if rating]
    return sum(ratings) / len(ratings) if ratings else None

df_final_with_places_covered['Average Hotel Rating'] = df_final_with_places_covered['Hotel Details'].apply(extract_average_rating)

# Displaying the first few rows with the new hotel features
df_final_with_places_covered[['Hotel Details', 'Number of Hotels', 'Average Hotel Rating']].head()

# Dropping the original 'Hotel Details' column
df_final_with_places_covered.drop('Hotel Details', axis=1, inplace=True)
```

```python
#find the number of rows with missing values for 'Average Hotel Rating'
df_final_with_places_covered['Average Hotel Rating'].isnull().sum()
```

```python
#fill the missing values with the mean rating
df_final_with_places_covered['Average Hotel Rating'].fillna(df_final_with_places_covered['Average Hotel Rating'].mean(), inplace=True)
```

```python
df_hotel=df_final_with_places_covered.copy()
```

Airline

```python
# Preparing the data for multi-label binarization for 'Airline'
# Splitting the airlines into a list of airlines per row
airlines_lists = df_start_city['Airline'].str.split('|')

# Initializing the MultiLabelBinarizer
mlb_airlines = MultiLabelBinarizer()

# Applying MultiLabelBinarizer to the airlines lists
airlines_encoded = mlb_airlines.fit_transform(airlines_lists)

# Creating a DataFrame from the encoded data
airlines_encoded_df = pd.DataFrame(airlines_encoded, columns=mlb_airlines.classes_, index=df_start_city.index)

# Prefixing the column names with "Airline_" to distinguish from other features
airlines_encoded_df.columns = ['Airline_' + col for col in airlines_encoded_df.columns]

# Displaying the shape of the new DataFrame and a few rows to understand its structure
airlines_encoded_df_shape = airlines_encoded_df.shape
airlines_encoded_df_head = airlines_encoded_df.head()

airlines_encoded_df_shape, airlines_encoded_df_head
```

```python
# Merging the encoded airline data back into the main dataset
df_airlines = df_start_city.join(airlines_encoded_df)

# Displaying the shape of the final dataset and a few rows to understand its structure after merging
df_airlines.shape
```

```python
df_airlines.drop('Airline', axis=1, inplace=True)
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
X = df.drop('Per Person Price', axis=1)
y = df['Per Person Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import numpy as np
#train the random forest regressor model
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)

#predict the values
y_pred = rf.predict(X_test)

#calculate the mean squared error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print('RMSE:', rmse)

#calculate the mean absolute error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print('MAE:', mae)

#calculate the r2 score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('R2:', r2)
```

```python
#train catboost regressor model
from catboost import CatBoostRegressor
cat = CatBoostRegressor(random_state=42)
cat.fit(X_train, y_train)

#predict the values
y_pred = cat.predict(X_test)

#calculate the mean squared error
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print('RMSE:', rmse)

#calculate the mean absolute error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print('MAE:', mae)

#calculate the r2 score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('R2:', r2)
```
