```python
import pandas as pd

# Load the dataset
file_path = 'makemytrip.csv'
df = pd.read_csv(file_path, error_bad_lines=False)

# Displaying the first few rows of the dataset
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Reverting back to the dataset and dropping specific columns as requested
columns_to_drop_extended = ['Uniq Id', 'Crawl Timestamp', 'Page Url', 
                            'Flight Stops', 'Meals', 'Initial Payment For Booking', 'Date Change Rules']

df_reduced = df.drop(columns_to_drop_extended, axis=1)
df_reduced.head()
```

```python
df_reduced.shape
```

```python
# Dropping all rows with missing values in the reduced dataset
df_reduced_cleaned = df_reduced.dropna()

df_reduced_cleaned_shape = df_reduced_cleaned.shape

df_reduced_cleaned_shape
```

```python
df_reduced_cleaned.isnull().sum()
```

## Feature engineeing


```python
# Analyzing the dataset for feature engineering and encoding categorical variables
df_reduced_cleaned.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generating a word cloud for the 'Package Name' column
text = ' '.join(df_reduced_cleaned['Package Name'])

# Creating a WordCloud object
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(text)

# Plotting the WordCloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Feature Engineering
# Extracting total nights from 'Itinerary'
df_reduced_cleaned['Total Nights'] = df_reduced_cleaned['Itinerary'].str.extractall(r'(\d+)N').astype(int).groupby(level=0).sum()

# Counting number of destinations
df_reduced_cleaned['Number of Destinations'] = df_reduced_cleaned['Destination'].apply(lambda x: len(x.split('|')))

# Counting number of sightseeing places
df_reduced_cleaned['Number of Sightseeing Places'] = df_reduced_cleaned['Sightseeing Places Covered'].apply(lambda x: len(x.split('|')))

# Encoding Categorical Variables
# Selecting categorical columns for encoding
categorical_cols = ['Package Type', 'Company', 'Start City']

# Applying one-hot encoding
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_cols = pd.DataFrame(encoder.fit_transform(df_reduced_cleaned[categorical_cols]), index=df_reduced_cleaned.index)

# Merging encoded columns back to the dataframe
df_encoded = df_reduced_cleaned.join(encoded_cols)

# Dropping original columns that were used for feature engineering and encoding
cols_to_drop = ['Itinerary', 'Destination', 'Sightseeing Places Covered'] + categorical_cols
df_final = df_encoded.drop(cols_to_drop, axis=1)

# Displaying the first few rows of the final dataset and the shape
df_final_head = df_final.head()
df_final_shape = df_final.shape

df_final_head, df_final_shape
```

```python
# Displaying the unique values in the 'Package Type' column
unique_package_types = df_reduced_cleaned['Package Type'].unique()
unique_package_types
```

```python
# Dropping rows where 'Package Type' contains URLs
# This is achieved by checking if 'Package Type' contains 'http' which is indicative of a URL
df_reduced_no_urls = df_reduced_cleaned[~df_reduced_cleaned['Package Type'].str.contains('http')]

# Displaying the unique values in the 'Package Type' column after removing URLs
unique_package_types_after_removal = df_reduced_no_urls['Package Type'].unique()
unique_package_types_after_removal, df_reduced_no_urls.shape
```

```python
import seaborn as sns

# Plotting the distribution of package types
plt.figure(figsize=(10, 6))
sns.countplot(data=df_reduced_no_urls, x='Package Type', order = df_reduced_no_urls['Package Type'].value_counts().index)
plt.title('Distribution of Package Types')
plt.xlabel('Package Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Displaying the unique values in the 'Company' column
unique_companies = df_reduced_no_urls['Company'].unique()
unique_companies
```

```python
# Dropping the 'Company' column as it contains only one unique value
df_reduced_no_company = df_reduced_no_urls.drop('Company', axis=1)

# Displaying the shape of the dataset after dropping the 'Company' column
df_reduced_no_company_shape = df_reduced_no_company.shape
df_reduced_no_company_shape
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Preparing the data for multi-label binarization
# Splitting the destinations into a list of destinations per row
destination_lists = df_reduced_no_company['Destination'].str.split('|')

# Initializing the MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Applying MultiLabelBinarizer to the destination lists
destination_encoded = mlb.fit_transform(destination_lists)

# Creating a DataFrame from the encoded data
destination_encoded_df = pd.DataFrame(destination_encoded, columns=mlb.classes_, index=df_reduced_no_company.index)

# Displaying the shape of the new DataFrame and a few rows to understand its structure
destination_encoded_df_shape = destination_encoded_df.shape
destination_encoded_df_head = destination_encoded_df.head()

destination_encoded_df_shape, destination_encoded_df_head
```

```python
df_destination_encoded = df_reduced_no_company.join(destination_encoded_df)
#dropping the original 'Destination' column
df_destination_encoded = df_destination_encoded.drop('Destination', axis=1)
```

```python
df_destination_encoded.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Extracting TF-IDF features from 'Package Name'
tfidf = TfidfVectorizer(max_features=50)  # Limiting to top 50 features for simplicity
tfidf_matrix = tfidf.fit_transform(df_reduced_no_company['Package Name'])

# Creating a DataFrame from the TF-IDF features
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out(), index=df_reduced_no_company.index)

# Displaying the shape of the TF-IDF DataFrame and a few rows to understand its structure
tfidf_df_shape = tfidf_df.shape
tfidf_df_head = tfidf_df.head()

tfidf_df_shape, tfidf_df_head
```

```python
df_tfidf = df_destination_encoded.join(tfidf_df)
# Dropping the original 'Package Name' column
df_tfidf = df_tfidf.drop('Package Name', axis=1)

df_tfidf.head()
```

```python
# Counting the number of different places/stops in each package's itinerary
# Assuming each stop in the itinerary is separated by a '.'
df_tfidf['Number of Stops in Itinerary'] = df_tfidf['Itinerary'].str.count('\.') + 1

# Displaying the first few rows of the updated dataset with the new feature
df_tfidf[['Itinerary', 'Number of Stops in Itinerary']].head()
```

```python
df_tfidf.drop('Itinerary', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Preparing the data for multi-label binarization for 'Places Covered'
# Splitting the places into a list of places per row
places_covered_lists = df_tfidf['Places Covered'].str.split('|')

# Initializing the MultiLabelBinarizer
mlb_places = MultiLabelBinarizer()

# Applying MultiLabelBinarizer to the places covered lists
places_covered_encoded = mlb_places.fit_transform(places_covered_lists)

# Creating a DataFrame from the encoded data
places_covered_encoded_df = pd.DataFrame(places_covered_encoded, columns=mlb_places.classes_, index=df_tfidf.index)

# Prefixing the column names with "Covered_" to distinguish from destinations
places_covered_encoded_df.columns = ['Covered_' + col for col in places_covered_encoded_df.columns]

# Displaying the shape of the new DataFrame and a few rows to understand its structure
places_covered_encoded_df_shape = places_covered_encoded_df.shape
places_covered_encoded_df_head = places_covered_encoded_df.head()

places_covered_encoded_df_shape, places_covered_encoded_df_head
```

```python
# Merging the encoded places covered data back into the main dataset
df_final_with_places_covered = df_tfidf.join(places_covered_encoded_df)

# Displaying the shape of the final dataset and a few rows to understand its structure after merging
df_final_with_places_covered_shape = df_final_with_places_covered.shape
df_final_with_places_covered_head = df_final_with_places_covered.head()

df_final_with_places_covered_shape, df_final_with_places_covered_head
```

```python
# Dropping the original 'Places Covered' column
df_final_with_places_covered.drop('Places Covered', axis=1, inplace=True)
```

```python
import datetime

# Converting 'Travel Date' to datetime
df_final_with_places_covered['Travel Date'] = pd.to_datetime(df_final_with_places_covered['Travel Date'], errors='coerce')

# Extracting year, month, and day
df_final_with_places_covered['Year'] = df_final_with_places_covered['Travel Date'].dt.year
df_final_with_places_covered['Month'] = df_final_with_places_covered['Travel Date'].dt.month
df_final_with_places_covered['Day'] = df_final_with_places_covered['Travel Date'].dt.day

# Extracting day of the week
df_final_with_places_covered['Day of Week'] = df_final_with_places_covered['Travel Date'].dt.dayofweek

# Displaying the first few rows of the dataset with the new date features
df_final_with_places_covered[['Travel Date', 'Year', 'Month', 'Day', 'Day of Week']].head()
```

```python
# Dropping the original 'Travel Date' column
df_final_with_places_covered.drop('Travel Date', axis=1, inplace=True)
```

```python
df_final_with_places_covered.shape
```

```python
import re

# Feature 1: Number of Hotels
# Counting the number of different hotels mentioned for each package
# Assuming each hotel is separated by a '|'
df_final_with_places_covered['Number of Hotels'] = df_final_with_places_covered['Hotel Details'].str.count('\|') + 1

# Feature 2: Average Hotel Rating
# Extracting hotel ratings and calculating the average rating
def extract_average_rating(hotel_details):
    # Finding all occurrences of hotel ratings (assumed to be in the format 'Hotel Name:Rating')
    ratings = re.findall(r':(\d\.\d)', hotel_details)
    ratings = [float(rating) for rating in ratings if rating]
    return sum(ratings) / len(ratings) if ratings else None

df_final_with_places_covered['Average Hotel Rating'] = df_final_with_places_covered['Hotel Details'].apply(extract_average_rating)

# Displaying the first few rows with the new hotel features
df_final_with_places_covered[['Hotel Details', 'Number of Hotels', 'Average Hotel Rating']].head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding for 'Start City'
label_encoder = LabelEncoder()
df_final_with_places_covered['Start City Encoded'] = label_encoder.fit_transform(df_final_with_places_covered['Start City'])

# Displaying the first few rows of the dataset with the new label encoded 'Start City'
df_final_with_places_covered[['Start City', 'Start City Encoded']].head()
```

```python
# Preparing the data for multi-label binarization for 'Airline'
# Splitting the airlines into a list of airlines per row
airlines_lists = df_final_with_places_covered['Airline'].str.split('|')

# Initializing the MultiLabelBinarizer
mlb_airlines = MultiLabelBinarizer()

# Applying MultiLabelBinarizer to the airlines lists
airlines_encoded = mlb_airlines.fit_transform(airlines_lists)

# Creating a DataFrame from the encoded data
airlines_encoded_df = pd.DataFrame(airlines_encoded, columns=mlb_airlines.classes_, index=df_final_with_places_covered.index)

# Prefixing the column names with "Airline_" to distinguish from other features
airlines_encoded_df.columns = ['Airline_' + col for col in airlines_encoded_df.columns]

# Displaying the shape of the new DataFrame and a few rows to understand its structure
airlines_encoded_df_shape = airlines_encoded_df.shape
airlines_encoded_df_head = airlines_encoded_df.head()

airlines_encoded_df_shape, airlines_encoded_df_head
```

```python
# Merging the encoded airline data back into the main dataset
df_final_with_airlines = df_final_with_places_covered.join(airlines_encoded_df)

# Displaying the shape of the final dataset and a few rows to understand its structure after merging
df_final_with_airlines_shape = df_final_with_airlines.shape
df_final_with_airlines_head = df_final_with_airlines.head()

df_final_with_airlines_shape, df_final_with_airlines_head
```

```python
df_final_with_airlines.drop('Airline', axis=1, inplace=True)
df_final_with_airlines.head()
```

```python
df_final_with_airlines.drop('Hotel Details', axis=1, inplace=True)
df_final_with_airlines.head()
```

```python
df_final_with_airlines.drop('Start City', axis=1, inplace=True)
df_final_with_airlines.head()
```

```python
df_final_with_airlines.drop('Cancellation Rules', axis=1, inplace=True)
```

```python
df_final_with_airlines.drop('Package Type', axis=1, inplace=True)
```

```python
df_final_with_airlines.drop('Onwards Return Flight Time', axis=1, inplace=True)
```

```python
df_final_with_airlines.drop('Price Per Two Persons', axis=1, inplace=True)
```

```python
df_final_with_airlines.drop('Sightseeing Places Covered', axis=1, inplace=True)
```

```python
#predict using random forest regressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

#calculate the mean squared error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print('MSE: ', mse)
```

```python
#predict using logistic regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

#calculate the mean squared error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print('MSE: ', mse)

#print actual and predicted values
y_test = y_test.reset_index(drop=True)
y_pred = pd.Series(y_pred)
df_compare = pd.concat([y_test, y_pred], axis=1)
df_compare.columns = ['Actual', 'Predicted']
df_compare.head()
```
